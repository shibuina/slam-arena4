```python
from astra_camera import Camera
import cv2
```

```python
cam = Camera()
```

```python
try:
    while True:
        depth, rgb = cam.get_depth_and_color()
        cv2.imshow('depth', depth)
        cv2.imshow('rgb', rgb)
        cv2.waitKey(cam.wait_time)
except:
    cam.unload()
```
